```python
import sys, os
import pandas as pd
```

the section below organizes the overlap file of chip-seq or DEG (coord_file) with EpiSig section (feature_file)
and generate a bed file 

    per episig cluster; per chip_type; per cancer high or low or all

also filter by number: if a cluster has too few overlapped features, we remove it (as function enrichment analysis should be biased)

```python
# for chip seq
sv_folder_base = "/stg3/data3/peiyao/work/20230922.Lina_lung/results/20240215.episig_spec_feature_functions/02.episig_spec_features"
chip_name = 'H3K4me1'
N_site_cutoff = 20


overlap_dt = pd.read_csv(os.path.join(
    '/stg3/data3/peiyao/work/20230922.Lina_lung/results/20240215.episig_spec_feature_functions/01.overlaps',
    'overlap_episig.{}.hg19.txt'.format(chip_name)), sep='\t', header=None, )
N = len(overlap_dt.columns)
overlap_dt = overlap_dt[overlap_dt[int(N-1)]>0]

for episig_cls_id, dt_ in overlap_dt.groupby(int(N-2)):
    # save data for every episig cluster
    sv_folder = os.path.join(sv_folder_base, 'EpiSig.cls_{}'.format(episig_cls_id))
    os.makedirs(sv_folder, exist_ok=True, )
    dt__ = dt_.copy()

    if len(dt__)>=N_site_cutoff:
        # save data to this folder
        # all
        # dt__[[0,1,2,3]].to_csv(os.path.join(sv_folder, 'all.txt'), sep='\t', header=None, index=None, )  
        
        dt__['pk_type'] = [int(i.split('_')[-1]) for i in list(dt__[3])]   
        # positive (cancer high)
        dt__high = dt__[dt__['pk_type']==1]
        # dt__high[[0,1,2,3]].to_csv(os.path.join(sv_folder, 'cancer_high.txt'), sep='\t', header=None, index=None, )  
        # negative (cancer low)
        dt__low = dt__[dt__['pk_type']==0]
        # dt__low[[0,1,2,3]].to_csv(os.path.join(sv_folder, 'cancer_low.txt'), sep='\t', header=None, index=None, )
    dt__low
```

```python
#for chip-seq
sv_folder_base = "/stg3/data3/peiyao/work/20230922.Lina_lung/results/20240215.episig_spec_feature_functions/02.episig_spec_features"

def proc_chip(chip_name, sv_folder_base=sv_folder_base):
    overlap_dt = pd.read_csv(os.path.join(
        '/stg3/data3/peiyao/work/20230922.Lina_lung/results/20240215.episig_spec_feature_functions/01.overlaps',
        'overlap_episig.{}.hg19.txt'.format(chip_name)), sep='\t', header=None, )
    overlap_dt = overlap_dt[overlap_dt[9]>0]
    for episig_cls_id, dt_ in overlap_dt.groupby(8):
        # save data for every episig cluster
        sv_folder = os.path.join(sv_folder_base, 'EpiSig.cls_{}'.format(episig_cls_id), chip_name)
        os.makedirs(sv_folder, exist_ok=True, )
        dt__ = dt_.copy()
        # save data to this folder
        # all
        dt__[[0,1,2,3]].to_csv(os.path.join(sv_folder, 'all.txt'), sep='\t', header=None, index=None, )  
        dt__['pk_type'] = [int(i.split('_')[-1]) for i in list(dt__[3])]   
        # positive (cancer high)
        dt__high = dt__[dt__['pk_type']==1]
        dt__high[[0,1,2,3]].to_csv(os.path.join(sv_folder, 'cancer_high.txt'), sep='\t', header=None, index=None, )  
        # negative (cancer low)
        dt__low = dt__[dt__['pk_type']==0]
        dt__low[[0,1,2,3]].to_csv(os.path.join(sv_folder, 'cancer_low.txt'), sep='\t', header=None, index=None, )
    return


```

```python
# iterate for chips

for chip_name in ['H3K27me3', 'H3K4me3', 'H3K27ac', 'H3K4me1', 'H3K9me3']:
    proc_chip(chip_name=chip_name, )
```

```python
# for DEG
chip_name = 'DEG'
overlap_dt = pd.read_csv(os.path.join(
    '/stg3/data3/peiyao/work/20230922.Lina_lung/results/20240215.episig_spec_feature_functions/01.overlaps',
    'overlap_episig.{}.hg19.txt'.format(chip_name)), sep='\t', header=None, )
overlap_dt = overlap_dt[overlap_dt[10]>0]
for episig_cls_id, dt_ in overlap_dt.groupby(9):
    # save data for every episig cluster
    sv_folder = os.path.join(sv_folder_base, 'EpiSig.cls_{}'.format(episig_cls_id), chip_name)
    os.makedirs(sv_folder, exist_ok=True, )
    dt__ = dt_.copy()
    # save data to this folder
    # all
    dt__[[0,1,2,3,4]].to_csv(os.path.join(sv_folder, 'all.txt'), sep='\t', header=None, index=None, )  
    # col 4 is log2FC (normal vs cancer), so + means normal high
    # positive (cancer high)
    dt__high = dt__[dt__[4]<0]
    dt__high[[0,1,2,3]].to_csv(os.path.join(sv_folder, 'cancer_high.txt'), sep='\t', header=None, index=None, )  
    # negative (cancer low)
    dt__low = dt__[dt__[4]>0]
    dt__low[[0,1,2,3]].to_csv(os.path.join(sv_folder, 'cancer_low.txt'), sep='\t', header=None, index=None, )
```

this section below organizes the overlap results between the cluster-spec chips (coord_file) with gene coord gencode file (feature_file)

```python
sv_folder_base = '/stg3/data3/peiyao/work/20230922.Lina_lung/results/20240215.episig_spec_feature_functions/03.episig_features_2_geneid_overlaps'
def analyze_overlap(episig_cls_id : int, chip_name : str, pk_type : str, sv_folder_base : str =sv_folder_base):
    
    dt_path = os.path.join(sv_folder_base, 'EpiSig.cls_{}/{}/{}.hg19.bed'.format(episig_cls_id, chip_name, pk_type))
    if os.path.exists(dt_path):
        dt = pd.read_csv(os.path.join(sv_folder_base, 'EpiSig.cls_{}/{}/{}.hg19.bed'.format(episig_cls_id, chip_name, pk_type)),
            sep='\t', header=None, 
        )
        if len(dt)>0:
            dt = dt[dt[8]>0]
            if len(dt)>0:
                return ' '.join(set(list(dt[7])))
    return ' '
```

```python
sv_file = os.path.join('/stg3/data3/peiyao/work/20230922.Lina_lung/results/20240215.episig_spec_feature_functions/04.epi_features_gid',
                       'gene_list.txt')
f = open(sv_file, 'w')
for episig_cls_id in range(1,7):
    for chip_name in ['H3K27me3', 'H3K4me3', 'H3K27ac', 'H3K4me1', 'H3K9me3']:
        for pk_type in ['all','cancer_high', 'cancer_low']:
            f.write('epi_{}.{}.{}:\n'.format(episig_cls_id, chip_name, pk_type))
            # print('epi_{}.{}.{}:\n'.format(episig_cls_id, chip_name, pk_type))
            f.write(analyze_overlap(episig_cls_id, chip_name , pk_type, ))
            # print(analyze_overlap(episig_cls_id, chip_name , pk_type, ))
            f.write('\n')
f.close()
```
